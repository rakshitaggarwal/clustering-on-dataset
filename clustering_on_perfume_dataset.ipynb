```python
from pycaret.utils import version
version()
```

```python
from pycaret.datasets import get_data
get_data('index')
```

```python
perfume_dataset=get_data('perfume')
```

```python
perfume_dataset.to_csv("perfume_dataset.csv")
from google.colab import files
files.download('perfume_dataset.csv')
```

```python
from pycaret.clustering import *
KMeansClusteringParameters=setup(perfume_dataset)
```

```python
#building kmeans clustering model
KMeansClusteringModel=create_model('kmeans',num_clusters=4)
```

```python
#assign model-assigning the labels to dataset
KMeansPrediction=assign_model(KMeansClusteringModel)
KMeansPrediction
```

```python
KMeansPrediction.to_csv("KMeansResult.csv")
print("Result file saved successfully")
```

```python
#saving the trained model
x=save_model(KMeansClusteringModel,'KMeansClusteringModelFile')
```

```python
#download the model
from google.colab import files
files.download('KMeansClusteringModelFile.pkl')
```

```python
#load the model
KMeansClusteringModel1=load_model('KMeansClusteringModelFile')
```

```python
#clustering new unseen dataset
#selecting top 10 rows
new_data=get_data('perfume').iloc[:10]
```

```python
new_predictions=predict_model(KMeansClusteringModel,data=new_data)
new_predictions
```

```python
#saving predicion result to csv
new_predictions.to_csv("NewPredictions.csv")
```

```python
#evaluating the model
evaluate_model(KMeansClusteringModel)
```
